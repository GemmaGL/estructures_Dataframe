De 200.000 vols 126.109 han arribat amb retard. El que suposa un 63% de tots els vols.

Per estudiar que ocasiona principalment els retards, es calcularà el total de minuts de retard total i per cada categoria:
- Carr: ocasionats per la companyia aèrea
- Weat: ocasionats per les condicions climatològiques
- NAS: ocasionats per la Agencia Nacioanal Espacial
- Sec: ocasionats per temes relacionats amb la Seguretat
- LAD: ocasionats per avions que venien d'altres destins amb retard.

```python
total_delay = df_delay["ArrDelay"].sum() 
total_delay_carr = df_delay["CarrierDelay"].sum()
total_delay_weat = df_delay["WeatherDelay"].sum()
total_delay_NAS = df_delay["NASDelay"].sum()
total_delay_Sec = df_delay["SecurityDelay"].sum()
total_delay_LAD = df_delay["LateAircraftDelay"].sum()
per_delay_carr = round((total_delay_carr / total_delay) * 100, 2)
per_delay_weat = round((total_delay_weat / total_delay)*100 , 2)
per_delay_NAS = round((total_delay_NAS / total_delay)*100 , 2)
per_delay_Sec = round((total_delay_Sec / total_delay)*100 , 2)
per_delay_LAD = round((total_delay_LAD / total_delay)*100 , 2)

print("Del total de minuts dels retards:{}\n -{}% CARR \n -{}% Weat\n -{}%  NAS\n -{}%  Seg \n -{}% LAD.".format(total_delay, per_delay_carr , per_delay_weat, per_delay_NAS ,
per_delay_Sec , per_delay_LAD))
```

Si ho refactoritzem quedaria així:

```python
columns = {
    "CarrierDelay": "CARR",
    "WeatherDelay": "Weat",
    "NASDelay": "NAS",
    "SecurityDelay": "Seg",
    "LateAircraftDelay": "LAD"
}

total_delay = df_delay["ArrDelay"].sum()
frequencies = {}

for column, label in columns.items():
    total_column_delay = df_delay[column].sum()
    percentage_delay = round((total_column_delay / total_delay) * 100, 2)
    frequencies[label] = percentage_delay

print("Del total de minuts dels retards: {}\n".format(total_delay))

for label, percentage in frequencies.items():
    print("- {}% són {}".format(percentage, label))
```
